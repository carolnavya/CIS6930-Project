# Layer 1 Model to generate BrainSpan prediction scores


```r
library(randomForest)
```

```r
#load pre-processed data files
load("rData/Genewise_expr_matrix.Rdata")
load("rData/training_labels.Rdata")

# all examples as input features
features = genewise_expr[ rownames(genewise_expr) %in% c(pos_genes,neg_genes) , ]
# only positive examples as lables
labels = as.factor(rownames(features) %in% pos_genes)

# train random forest
set.seed(5136)
brainspan_model = randomForest(x=features,
                     y=labels,
                     ntree=1000,
                     importance=T,
                     strata=labels,
                     sampsize=c(76,76),
                     do.trace=10,
                     proximity=T)

# get predictions on all remaining genes
other_preds =  na.roughfix(genewise_expr[ !rownames(genewise_expr) %in% rownames(features) , ])

bs_preds = predict(brainspan_model, other_preds, type="prob")
bs_preds = rbind(brainspan_model$vote, bs_preds)

# save model
save(brainspan_model, bs_preds, file="models/brainspan_score.Rdata")

```
